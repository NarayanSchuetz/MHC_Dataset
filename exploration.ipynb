```python
info_df = pd.read_parquet("/Users/narayanschuetz/tmp_data/mhc_full_participant_info.parquet")
# Convert birthdate and createdOn to datetime
info_df['birthdate'] = pd.to_datetime(info_df['birthdate'], errors='coerce', utc=True)
info_df['createdOn'] = pd.to_datetime(info_df['createdOn'], errors='coerce', utc=True)

# Calculate age at the time the record was created
info_df['age'] = (info_df['createdOn'] - info_df['birthdate']).dt.days / 365.25
info_df['age'] = info_df['age'].round().astype('Int64')  # Round to nearest integer and use nullable integer type

# Display the birthdate and calculated age
info_df[['birthdate', 'age']].head(10)
```

```python
df = pd.read_parquet("/Users/narayanschuetz/tmp_data/mhc_demographics.parquet")

df = df.merge(info_df, on="healthCode", how="left")

# Create discrete bins for weight and height
# First, let's handle the weight (in kg)
weight_bins = [0, 50, 60, 70, 80, 90, 100, float('inf')]
weight_labels = ['<50kg', '50-60kg', '60-70kg', '70-80kg', '80-90kg', '90-100kg', '>100kg']

# Add a category for unknown values
df['weight_group'] = pd.cut(
    df['WeightKilograms'], 
    bins=weight_bins, 
    labels=weight_labels, 
    right=False
)

# Convert to string type to allow adding the "Unknown" category
df['weight_group'] = df['weight_group'].astype(str)
df.loc[df['weight_group'] == 'nan', 'weight_group'] = 'Unknown_Weight'

# Now, let's handle the height (in cm)
height_bins = [0, 150, 160, 170, 180, 190, 200, float('inf')]
height_labels = ['<150cm', '150-160cm', '160-170cm', '170-180cm', '180-190cm', '190-200cm', '>200cm']

# Create a new column with the binned height values
df['height_group'] = pd.cut(
    df['HeightCentimeters'], 
    bins=height_bins, 
    labels=height_labels, 
    right=False
)

# Convert to string type to allow adding the "Unknown" category
df['height_group'] = df['height_group'].astype(str)
df.loc[df['height_group'] == 'nan', 'height_group'] = 'Unknown_Height'

# Display the distribution of the binned values
print("Weight Group Distribution:")
print(df['weight_group'].value_counts())

print("\nHeight Group Distribution:")
print(df['height_group'].value_counts())

# Now, let's handle the age
age_bins = [18, 25, 35, 45, 55, 65, 75, float('inf')]
age_labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65-74', '75+']

# Create a new column with the binned age values
df['age_group'] = pd.cut(
    df['CurrentAge'], 
    bins=age_bins, 
    labels=age_labels, 
    right=False
)

# Convert to string type to allow adding the "Unknown" category
df['age_group'] = df['age_group'].astype(str)
df.loc[df['age_group'] == 'nan', 'age_group'] = 'Unknown_Age'

# Display the distribution of the binned age values
print("\nAge Group Distribution:")
print(df['age_group'].value_counts())
```
